# Import data from web scraping

ในส่วนนี้ จะทำ Web scraping เพื่อดึงข้อมูล จากเว็ปไซต์ [Agnos forum](https://www.agnoshealth.com/forums/search) จะดึงโดยใช้ library ที่เขียนขึ้นเองตอนเรียนวิชา Data engineering ตอนเรียนปี 3 ครับ สามารถ clone ได้ที่ [Github Repo](https://github.com/Mahmimi/WebCrawler.git)

```python
from WebCrawler.WebScraping_Crawler import SinglePage_WebCrawler, MultiPage_WebCrawler
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

# Set the path to your ChromeDriver
service = Service(executable_path='./chromedriver-win64/chromedriver.exe')

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("start-maximized")
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

# Pass the ChromeDriver path to the webdriver
driver = webdriver.Chrome(service=service, options=options)

driver.implicitly_wait(10)
```

```python
url = "https://www.agnoshealth.com/forums/search?page={0}"
crawler = MultiPage_WebCrawler(url=url, 
                                html_check=False, 
                                category="healthcare", 
                                owner_source="Agnos Health",
                                driver=driver)
```

โดยจะ scrape เว็ปไซต์ทั้งหมดจำนวน 174 หน้า จำนวนรวมกระทู้ทั้งหมด 563 กระทู้ คัดเอาเฉพาะกระทู้ที่มีคุณหมอตอบเท่านั้น

```python
web_tag = {
                'url' : url,
                'start_page' : 1,
                'end_page' : 174,
                'step_page' : 1,
                'article_area_class_in_PageList': "flex flex-row justify-between mt-2 gap-4",
                'a_class_in_PageList' : 'undefined', 
                'title_tag_in_ArticlePage' : 'span',
                'title_class_in_ArticlePage' : 'font-bold text-lg',
                'content_area_tag_in_ArticlePage':'p',
                'content_area_class_in_ArticlePage':'mt-4',
                'banner_tag_in_ArticlePage':'src',
                'banner_class_in_ArticlePage':'',
                'custom_pageindex_list':None
            }
content = crawler.get_articles(web_tag)
```

# Export data to CSV

ในส่วนนี้จะทำการแปลงข้อมูลให้เป็น dataframe พร้อมทำการ preprocess คือเปลี่ยนชื่อคอลัมน์และตัดคอลัมน์ที่ไม่จำเป็นก่อนที่จะ export เป็น csv เพื่อให้ง่ายต่อการใช้งานต่อไปครับ 

```python
import pandas as pd

data = pd.DataFrame(content)
data.rename(columns={'title': 'Question', 'content': 'Answer'}, inplace=True)
data = data[["url", "Question", "Answer"]]
```

จะทำการ cleaning data เบื้องต้นคือลบเครื่องหมายวรรคตอน ทั้งหมดก่อน export เป็น CSV

```python
data['Question'] = data['Question'].str.strip()
data['Answer'] = data['Answer'].str.strip()
```

```python
data.to_csv("Agnos_Healthcare.csv", index=False, encoding='utf-8')
```
